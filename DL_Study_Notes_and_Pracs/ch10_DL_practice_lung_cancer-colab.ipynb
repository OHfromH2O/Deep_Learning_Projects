# 10장 딥러닝 모델 설계하기  (2장 딥러닝의 핵심 미리 보기)

### 1. 환경 준비

```python
# 텐서플로 라이브러리 안에 있는 케라스 API에서 필요한 함수들을 불러옵니다.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 데이터를 다루는 데 필요한 라이브러리를 불러옵니다.
import numpy as np
```

### 2. 데이터 준비

```python
# 깃허브에 준비된 데이터를 가져옵니다.
!git clone https://github.com/taehojo/data.git

# 준비된 수술 환자 데이터를 불러옵니다.
Data_set = np.loadtxt("./data/ThoraricSurgery3.csv", delimiter=",")
X = Data_set[:,0:16]    # 환자의 진찰 기록을 X로 지정합니다.
y = Data_set[:,16]      # 수술 1년 후 사망/생존 여부를 y로 지정합니다.
```

### 3. 구조 결정

```python
# 딥러닝 모델의 구조를 짜보자.
#Sequential 함수를 선언한 후, model.aad()라는 라인으로 층을 추가.
#중간의 은닉층 = 노드 30개, 입력데이터 16개를 가져온다, 활성화 함수는 ReLu
#마지막 층은 출력층이 된다: 출력 값은 하나로 나와야 하므로 노드가 1개이다, 최종 출력 활성화 함수는 시그모이드
model = Sequential()
model.add(Dense(30, input_dim=16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

### 4. 모델 컴파일 및 실행

```python
# 앞서 지정한 모델이 효과적으로 구동될 수 있도록 환경 설정 = 모델 컴파일
# 폐암 수술 환자의 생존율 예측 결과는 1, 0임
# 따라서, 우리는 손실 함수로 교차 엔트로피 함수를 사용, 경사하강법은 adam, 결과는 학습셋 정확도로 출력
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# model.fit() 함수로 정해진 환경에 따라 데이터를 불러서 실행.
# epoch = 학습 프로세스가 모든 샘플에 대해 한 번 실행되는 것
# epoch = 5 => 샘플이 처음부터 끝까지 5번 사용될 때까지 실행을 반보하겠다.
# batch_size = 샘플을 한 번에 몇개씩 처리할 지의 단위 (e.g.,470개의 샘플을 16개씩 끊어서 보겠다)
## 배치 사이즈가 너무 크면, 학습 속도가 느려지며; 너무 작으면 각 실행 값에 편차가 생겨 전체 결괏값이 불안해진다
history=model.fit(X, y, epochs=5, batch_size=16)
```
