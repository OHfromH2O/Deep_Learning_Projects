1.1.) **퍼셉트론** = 자동판독기. 입력된 신호들을 받아서 YES (1) or NO (0)을 결정하는 것.

1.2.)왜 최초의 학습인가?

-기존 컴퓨터 = 사람이 모든 규칙을 알려주어야 했었음.

-퍼셉트론 = 입력값을 여러개 받아 출력을 만들 때, 이 입력값에 **가중치 (w)를 스스로 고치는 능력**이 있는 것임.

-예를 들어, 랜덤 가중치를 두어서 시도 (고양이 사진 보고 개) -> 오차 발견 -> 가중치 수정 -> 과정 반복 -> 기계가 데이터로 규칙을 스스로 학습!!!


-이 경우, 학습 과정을 거치면서 (1,0)을 만족시키는 선이 여러 개 존재할 것임. (p84)

-**아달라안**: 여기에 경사하강법이 등장하면서, 최적의 선 하나만을 찾을 수 있게 됨.

+)**가중합**이란, 입력값과 가중치를 모두 곱한 후 바이어스를 더한 값 (활성화 함수 개념)

-아달라안을 기반으로 SVM (support vector machine)등 머신 러닝의 중요한 알고리즘들이 발전함.

3.1.)이를 해결하기 위해 등장한 개념이 XOR (exclusive OR)

-AND = 두 개 모두 1일때 1

-OR = 두 개 중 한 개라도 1이면, 1

-XOR = 하나만 1일 때, 1

이러한 AND, OR, XOR의 진리표를 좌표에 표현했을 때, 이 검정 포인트/흰색 포인트는 AND, OR에서는 직선 하나로 쉽게 구분할 수 있음.

그러나, XOR은 선을 그어 구분할 수 없는 한계가 있었음.

7.1.)이렇게 **은닉층**을 만들어서 다층 퍼셉트론 문제는 해결하였다. 그러나, 은닉층에 있는 가중치를 데이터를 통해 학습시키는 방법은 해결되지 않았다.

=> 이 방법이 거꾸로가서 은닉층의 가중치를 수정하는 **오차 역전파** 방법이다. (back propagation)

7.2.)오차 역전파의 개념

-(1) 순전파가 일어난다. 각 가중치의 초깃값이 정해진다. 이 가중치로 만들어진 값과 실제 값을 비교해 출력층의 오차를 계산할 수 있다.

-(2)이 출력층의 오차를 최소화하는 것이 우리의 목적이다. 이를 위해, 역으로 첫 번째 가중치 (w31)을 수정하고, 두 번째 가중치 (w11)을 수정하는 과정이 필요하다.

-(3)가중치를 업데이트하는 방법은, 오차 공식을 구하고, 해당 가중치 (w31)에 대해 편미분하면 된다.

-(4)그러나 **은닉층은 겉으로 드러나지 않으므로, 오차를 구할 만한 출력 값이 없다.!!**

-(5)이를 해결하기 위해, 출력층의 오차를 이용하여, 각 가중치를 업데이트하는 공식을 수립할 수 있다. **델타식 out(1-out)**

---

8.1.)델타식과 기울기 소실 문제 -> 새로운 활성화 함수의 등장

-델타식을 통해 우리는, 은닉층의 숫자를 늘려도 깊은 층의 계산을 계속 수행할 수 있게 되었다. 왜냐하면, 은닉층의 숫자가 늘어도, out(1-out) 형태는 계속 나타나기 때문이다.

-그러나 은닉층이 깊어질수록, 역전파를 진행하면, 기울기가 소실되는 문제가 나타난다. (시그모이드 활성화 함수의 한계)

-이를 해결하고자, **ReLU, 하이퍼볼릭 탄젠트, 소프트플러스**와 같은 새로운 활성화 함수들을 정의하여, 기울기 소실 문제를 해결하였다.

---

9.1)속도와 정확도 향상을 위한 고급 경사 하강법
(**옵티마이저의 발전**)

-기존의 경사 하강법은 한 번 업데이트할 때 마다, 전체 데이터를 미분하므로 속도가 느리고, 최적화 과정이 멈출 수 있다.

-확률적 경사 하강법 (Stochastic Gradient Descent, SGD) = 전체 데이터가 아니라, 랜덤 추출한 일부 데이터만 사용하기에 더 빠르고 자주 업데이트할 수 있다!!

-모멘텀 확률적 경사 하강법 (momentum SGD) = 오차를 수정 하기 전, 바로 앞 수정 방향 (+,-)를 참고하여 같은 방향으로만 비율이 수정되도록하는 방법이다. (같은 방향으로 관성을 주는 것이다). 이를 통해, 방향을 바꾸는 지그재그 현상이 줄어들도, 관성효과를 낼 수 있다. (더 빠르게 전진)

-최종적으로 현재, 이러한 속도와 정확도를 모두 향상 시킨, **아담 (adam)**이라는 고급 경사 하강법을 사용한다.

### 4. 코딩으로 XOR 문제 해결하기

```python
import numpy as np

# 가중치와 바이어스
w11 = np.array([-2, -2])
w12 = np.array([2, 2])
w2 = np.array([1, 1])
b1 = 3
b2 = -1
b3 = -1

# 퍼셉트론
def MLP(x, w, b):
    y = np.sum(w * x) + b
    if y <= 0:
        return 0
    else:
        return 1

# NAND 게이트
def NAND(x1,x2):
    return MLP(np.array([x1, x2]), w11, b1)

# OR 게이트
def OR(x1,x2):
    return MLP(np.array([x1, x2]), w12, b2)

# AND 게이트
def AND(x1,x2):
    return MLP(np.array([x1, x2]), w2, b3)

# XOR 게이트
def XOR(x1,x2):
    return AND(NAND(x1, x2),OR(x1,x2))


# x1, x2 값을 번갈아 대입해 가며 최종 값 출력
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = XOR(x[0], x[1])
    print("입력 값: " + str(x) + " 출력 값: " + str(y))
```
